#Description of the target audience of FreeCodeCamp E-learning products(Personas) 

In this document, I want to analyze a survey of prospective customers of a firm that organizes online IT courses. For a better presentation of the information, I will summarize all the results in the form of a technique used in UX (User Experience) **Personas**. 

[A persona, (also user persona, customer persona, buyer persona) in user-centered design and marketing is a fictional character created to represent a user type that might use a site, brand, or product in a similar way. Marketers may use personas together with market segmentation, where the qualitative personas are constructed to be representative of specific segments. The term persona is used widely in online and technology applications as well as in advertising, where other terms such as pen portraits may also be used.](/https://en.wikipedia.org/wiki/Persona_(user_experience))

This technique is useful for describing the target audience, when more than one attribute is needed for the description. In spite of the brevity and colorfulness of describing the target audience with Personas, according to some scientific studies, this method cannot authentically show the real customer. For a clearer picture, focus groups or interviews with potential customers are used. Since we already have a survey of the current target audience for our company, it seems to me, it will not be a mistake to use this method to form a description of the target audience.

##Data Cleaning

A huge number of fields have no value in themselves, and there is a high probability of having outliers (which can be seen in the table that aggregates the data). My goal in this part of the notebook is to clear the dataset of unnecessary data

```python
frc.head() #Look on top again
```

```python
frc1 = frc[['Age', 'ChildrenNumber', 'CountryCitizen', 'CountryLive', 'EmploymentField', 'EmploymentStatus', 'ExpectedEarning', 'HasChildren', 'Gender', 'HoursLearning', 'Income', 'IsUnderEmployed', 'JobPref', 'JobWherePref', 'LanguageAtHome', 'MaritalStatus', 'MoneyForLearning', 'SchoolDegree']]
print(frc1.columns) #Creating new dataset with needed columns only
```

```python
print(frc1.dtypes)
```

```python
#Chnage datetype of needed float columns to integer

frc1['HasChildren'] = frc1['HasChildren'].fillna(0.0).astype(int) 
frc1['Age'] = frc1['Age'].fillna(0.0).astype(int)
frc1['Income'] = frc1['Income'].fillna(0.0).astype(int)
frc1['ChildrenNumber'] = frc1['ChildrenNumber'].fillna(0.0).astype(int)
frc1['IsUnderEmployed '] = frc1['IsUnderEmployed'].fillna(0.0).astype(int)
frc1['MoneyForLearning'] = frc1['MoneyForLearning'].fillna(0.0).astype(int)
frc1['ExpectedEarning'] = frc1['ExpectedEarning'].fillna(0.0).astype(int)
```

###**Persona**

1.   *Physical data*

A middle-aged man (27-28 years old)
2.   *Employment Status*

Works in programming and IT, most likely has sufficient programming experience. Holds a bachelor's degree. Feels unappreciated in current job. Ideal situation is to be self-employed or work for a mid-sized company, freelance. Place of work is not so important, but has a preference for office work. 
3.   *Marital Status*

Single, never been married. In first-world countries (Canada, Australia, USA), this is the man among those surveyed who earns more. Married men mostly have one child and their income is not as high as that of single men. 

4.   *Geographical Status*

Most of the respondents live in the United States, with India in second place and Europe in third. They are most likely to speak English, Spanish, Hindi, and Russian. 



This is a short analysis of the information from the analysis above. For a more accurate description of the persona, I will try to describe the needs of such a client. 

The direction of programming is quite interesting now, so a lot of people who want to work for themselves are learning it. From the analysis of this survey, I realized that the main need of the respondents is not to change the field of activity (as most of them still work in the field of development), but to increase their knowledge and price in the market. A person does not necessarily want to leave their current job (this is indicated by a not so big difference between the pleasure and displeasure of working at their current place). 

Rather, people who learn using the firm's products are careerists (most have no family or only one child) who build up their knowledge and capital for well-being after age 30-40 (the average age of those surveyed, I repeat, is 27). 

Nevertheless, there is a not insignificant percentage of those who want to change the field to development. The best option, therefore, is to create a marketing strategy for two groups of the target audience 

*   Those who are expanding their knowledge and increasing their price in the marketplace (Most likely these people know exactly what they want, they already have goals and areas to explore) 
*   Those who are looking for a new job, they have high expectations and a lot of energy to learn new things (But, it's worth considering the burnout that can come during the course)
